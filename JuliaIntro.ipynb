# Differential Equations library
If not already installed, use the Pkg function. We'll also get a plotting library.

These will take a few minutes to install. It will also take a few minutes the first time you load the library.

```julia
using Pkg
Pkg.add("DifferentialEquations")
Pkg.add("Plots")
```

```julia
using DifferentialEquations
f(u,p,t) = 1.01*u
u0 = 1/2
tspan = (0.0,1.0)
prob = ODEProblem(f,u0,tspan)
sol = solve(prob, Tsit5(), reltol=1e-8, abstol=1e-8)

using Plots
plot(sol,linewidth=5,title="Solution to the linear ODE",
     xaxis="Time (t)",yaxis="u(t) (in μm)",label="Model") # legend=false
plot!(sol.t, t->0.5*exp(1.01t),lw=3,ls=:dash,label="True Solution!")
```
